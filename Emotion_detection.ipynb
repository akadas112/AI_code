```
from numpy import array
from google.colab import drive
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import  pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Embedding, Dense, LSTM, Bidirectional
import numpy as np
 
 
drive.mount('/content/drive')
 
def listToString(s):  
    
    # initialize an empty string 
    str1 = " " 
    
    # return string   
    return (str1.join(s)) 
# Get a list of keys from dictionary which has the given value
 
def getKeysByValue(dictOfElements, valueToFind):
    listOfKeys = list()
    listOfItems = dictOfElements.items()
    for item  in listOfItems:
        if item[1] == valueToFind:
            listOfKeys.append(item[0])
    for key  in listOfKeys:
        a = key
    return  a
 
def read_corpus(corpus_file):
    documents = []
    labels = []
    with open(corpus_file, encoding='utf-8') as f:
        for line in f:
            tokens = line.strip().split()
            documents.append(tokens[1:])
            labels.append( tokens[0] )
    for i in range(len(documents)):
        documents[i] = listToString(documents[i])
 
    emotions = list(dict.fromkeys(labels))
    class_type = { i : emotions[i] for i in range(0, len(emotions) ) }
    listOfLebels = list()
    for i in range(len(labels)):
        listOfLebels.append(getKeysByValue(class_type, labels[i]))
 
 
    return documents, labels, emotions, listOfLebels
 
 
X, Y, emotions, listOfLebels = read_corpus('/content/drive/My Drive/Dataset/train.txt')
```

```
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

# tokenization and converting words into sequence
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
dense_train_doc = tokenizer.texts_to_sequences(X)
print(dense_train_doc)

# padding the training documents in order to make them equal length.
MAX_LENGTH =2400

padded_trai_doc = pad_sequences(dense_train_doc, maxlen=MAX_LENGTH, padding='post')

print(padded_trai_doc)
print(Y)
#print(listOfLebels)

y=np.array(Y)
print(y)


encoder=OneHotEncoder(sparse=False)
y=y.reshape((4700, 1))

y1=encoder.fit_transform(y)


# Model Declaration
VOCUB_SIZE = 20000
model = Sequential()

# embedding layer
embedding_layer = Embedding(VOCUB_SIZE, 32, input_length=MAX_LENGTH)
model.add(embedding_layer)

# Flattening Layer
# model.add(Flatten())

# LSTM Extra - for performance
model.add(LSTM(units=128))

# Bidirectional LSTM Extra
# forward_layer = LSTM(128, return_sequences=False)
# backward_layer = LSTM(128, return_sequences=False, go_backwards=True)
# model.add(Bidirectional(forward_layer, backward_layer=backward_layer))

# output layer
model.add(Dense(units=6, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='Rmsprop', metrics=['accuracy'])

print(model.summary())
#y=np.array(listOfLebels)
#y1=np.reshape(y, (4700,1))
#y=np.array(listOfLebels)
#print(padded_trai_doc.shape)
#print(y1.shape)
#model.fit(padded_trai_doc, Y, epochs=100, verbose=1)
model.fit(padded_trai_doc, y1,epochs=5)
```
